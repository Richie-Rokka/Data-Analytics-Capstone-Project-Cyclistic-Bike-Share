```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
file_paths = glob.glob("C:/Data Analytics/Bike/*.csv")
df_list = [pd.read_csv(file) for file in file_paths]
df = pd.concat(df_list, ignore_index=True)
                       
                       
```

```python
bike_trips = df
```

```python
# Remove lat, long
# The 'errors="ignore"' argument prevents an error if a column is not found
bike_trips = bike_trips.drop(columns=['start_lat', 'start_lng', 'end_lat', 'end_lng'], errors='ignore')
```

```python
bike_trips.info()
```

```python
# Convert 'started_at' and 'ended_at' to datetime objects
bike_trips['started_at'] = pd.to_datetime(bike_trips['started_at'])
bike_trips['ended_at'] = pd.to_datetime(bike_trips['ended_at'])
```

```python
# Add columns that list the date, month, and day of each ride
bike_trips['date'] = bike_trips['started_at'].dt.date
bike_trips['month'] = bike_trips['started_at'].dt.month_name()
bike_trips['day'] = bike_trips['started_at'].dt.day
bike_trips['day_of_week'] = bike_trips['started_at'].dt.day_name()
```

```python
# Add a "ride_length" calculation to all_trips (in seconds)
bike_trips['ride_length'] = (bike_trips['ended_at'] - bike_trips['started_at']).dt.total_seconds()
```

```python
bike_trips.info()
```

```python
bike_trips.head()
```

```python
# Descriptive analysis on ride_length (all figures in seconds)
print("\nDescriptive statistics for ride_length:\n", bike_trips['ride_length'].describe())
```

```python
# Compare members and casual users
print("\nRide length statistics grouped by member_casual:\n", bike_trips.groupby('member_casual')['ride_length'].agg(['mean', 'median', 'max', 'min']))
```

```python
bike_trips.drop(bike_trips[bike_trips['member_casual'] == '60'].index, inplace=True)
```

```python
print("\nRide length statistics grouped by member_casual:\n", bike_trips.groupby('member_casual')['ride_length'].agg(['mean', 'median', 'max', 'min']))
```

```python
counts = bike_trips.groupby('member_casual')['ride_length'].agg(['mean', 'median', 'max', 'min']).reset_index()
counts.to_csv('Ride_Length_Statistics.csv', index=False)
```

```python
# See the average ride time by each day for members vs casual riders
print("\nAverage ride length by member type and day of the week:\n", bike_trips.groupby(['member_casual', 'day_of_week'])['ride_length'].mean())
```

```python
# Order the days of the week for correct sorting for analysis
days_order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
bike_trips['day_of_week'] = pd.Categorical(bike_trips['day_of_week'], categories=days_order, ordered=True)
```

```python
print("\nAverage ride length (sorted by day of week):\n", bike_trips.groupby(['member_casual', 'day_of_week'])['ride_length'].mean())
```

```python
counts = bike_trips.groupby(['member_casual', 'day_of_week'])['ride_length'].mean().reset_index()
counts.to_csv('Average_ride_length.csv', index=False)
```

```python
# Analyze rider ride length and number of rides data by type and weekday
summary_stats = bike_trips.groupby(['member_casual', 'day_of_week']).agg(number_of_rides=('ride_id', 'count'),
    average_duration=('ride_length', 'mean')).reset_index().round(2)
```

```python
print("\nSummary of rides and duration by rider type and weekday:\n", summary_stats)
```

```python
counts = bike_trips.groupby(['member_casual', 'day_of_week']).agg(number_of_rides=('ride_id', 'count'),
    average_duration=('ride_length', 'mean')).reset_index().round(2)
counts.to_csv('Daily_Ride_Summary.csv', index=False)
```

```python
summary_stats2 = bike_trips.groupby(['member_casual', 'month']).agg(number_of_rides=('ride_id', 'count'),
    average_duration=('ride_length', 'mean')).reset_index().round(2)
```

```python
print("\nSummary of rides and duration by rider type and month:\n", summary_stats2)
```

```python
counts = bike_trips.groupby(['member_casual', 'month']).agg(number_of_rides=('ride_id', 'count'),
    average_duration=('ride_length', 'mean')).reset_index().round(2)
counts.to_csv('Monthly_ride_Summary.csv', index=False)
```
